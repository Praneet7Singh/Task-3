```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

LOADING DATA

```python
data=pd.read_csv("IRIS.csv")
data.head()
```

DATA PREPROCESSING

```python
data["species"].unique()
```

```python
data.info()
```

```python
data.describe(include="all")
```

```python
data=data.dropna(axis=0)
```

```python
data=data.drop_duplicates(keep="first")
data
```

VISUALIZATION

```python
data.plot()
```

```python
f,(ax1,ax2,ax3)=plt.subplots(1,3,sharey=True,figsize=(15,3))
ax1.scatter(data["sepal_length"],data["species"])
ax1.set_title("species and sepal_length")
ax2.scatter(data["sepal_width"],data["species"])
ax2.set_title("species and sepal_width")
ax3.scatter(data["petal_length"],data["species"])
ax3.set_title("species and petal_length")
plt.show()
```

TRAINING DATA AND SCALING

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x=data.drop("species",axis=1)
y=data["species"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

ML MODEL

```python
KN_classifier = KNeighborsClassifier(n_neighbors=3)
KN_classifier.fit(x_train,y_train)
```

PREDICTION AND ACCURACY

```python
prediction=KN_classifier.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, prediction)
accuracy
```

CONFUSION MATRIX AND CLASSIFICATION REPORT

```python
confusion_matrix(y_true=y_test,y_pred=prediction)
```

```python
print(classification_report(prediction,y_test))
```
